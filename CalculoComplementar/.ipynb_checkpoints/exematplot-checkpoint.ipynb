
## Exercício 1

**Siga estas etapas:**
* **Crie um objeto de figura chamado fig usando plt.figure ()**
* **Use add_axes para adicionar um eixo à tela da figura em [0,0,1,1]. Chame esse novo eixo de eixo.**
* **Plote (x, y) nesses eixos e defina os rótulos e títulos para coincidir com o gráfico abaixo:**

```python
import matplotlib.pyplot as plt 
import numpy as np 

x = np.linspace(0,100,6)
y = np.array([0,50, 100, 150, 200])

fig = plt.figure()
eixo = fig.add_axes([0.1, 0.1, 0.8, 0.8])
eixo.plot(x, y, 'b')
eixo.set_xlabel('X')
eixo.set_ylabel('Y')
eixo.set_title('title')
plt.show()
```


## Exercício 2
**Crie um objeto de figura e coloque dois eixos nele, ax1 e ax2. Localizado em [0,0,1,1] e [0,2,0,5, .2, .2], respectivamente.**

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2, 0.5, .2, .2])
```


**Agora plote (x, y) nos dois eixos. E chame seu objeto de figura para mostrá-lo.**

```python
ax1.plot(x, y, 'r')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('no title')
ax2.plot(x, y, 'r')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
```

## Exercício 4

**Use plt.subplots (nrows = 1, ncols = 2) para criar o gráfico abaixo.**

```python
fig = plt.subplots(nrows=1, ncols=2)
```

**Agora plote (x, y) e (x, z) nos eixos. Brinque com a largura de linha e o estilo**

```python
x = np.linspace(0,100,6)
y = np.linspace(0,200,6)
z = np.linspace(0,10000, 6)


fig, ax1 = plt.subplots(figsize=(2.3, 4))
fig, ax2 = plt.subplots(figsize=(2.3, 4))
ax1.plot(x, y, color='black', lw=4, ls='--', markersize=8)
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.plot(x, z, 'r')
ax2.set_xlabel('x')
ax2.set_ylabel('z')
```


**Veja se você pode redimensionar o gráfico adicionando o argumento figsize () em plt.subplots () está copiando e colando o código anterior.**

```python
x = np.linspace(0,100,6)
y = np.linspace(0,200,6)
z = np.linspace(0,10000, 6)


fig, ax1 = plt.subplots(figsize=(5, 2.3))
fig, ax2 = plt.subplots(figsize=(5, 2.3))
ax1.plot(x, y, color='black', lw=4, ls='-', markersize=8)
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.plot(x, z, color='red', lw=3, ls='--')
ax2.set_xlabel('x')
ax2.set_ylabel('z')
```
